```python
#generate ORF

from Bio import Entrez

from Bio import SeqIO

Entrez.email='francisrobitaille07@gmail.com'

#mit codons produce a lot more ORFs than cyt ones. (more start codons, less stop ones)
cytstop = ['UAA','UAG','UGA']
cytstart = ['AUG']
mitstop = ['UAA','UAG'] #add tRNA molecular pattern + ribosome stalling patterns?
mitnouvstop = ['AGA','AGG']
mitstart = ['AUG','AUU','AUA'] #is there molecular evidence, or could it point to some other initiation mechanism?

#replace Z with real AA value 
mit = {"UUU": "F", "UUC": "F", "UUA": "L", "UUG": "L",
    "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "AUU": "I", "AUC": "I", "AUA": "I", "AUG": "M",
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "UAU": "Y", "UAC": "Y", "UAA": "*", "UAG": "*",
    "CAU": "H", "CAC": "H", "CAA": "Q", "CAG": "Q",
    "AAU": "N", "AAC": "N", "AAA": "K", "AAG": "K",
    "GAU": "D", "GAC": "D", "GAA": "E", "GAG": "E",
    "UGU": "C", "UGC": "C", "UGA": "Z", "UGG": "W",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R",
    "AGU": "S", "AGC": "S", "AGA": "R", "AGG": "R",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G"}

mit_nouv = {"UUU": "F", "UUC": "F", "UUA": "L", "UUG": "L",
    "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "AUU": "I", "AUC": "I", "AUA": "I", "AUG": "M",
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "UAU": "Y", "UAC": "Y", "UAA": "Z", "UAG": "Z",
    "CAU": "H", "CAC": "H", "CAA": "Q", "CAG": "Q",
    "AAU": "N", "AAC": "N", "AAA": "K", "AAG": "K",
    "GAU": "D", "GAC": "D", "GAA": "E", "GAG": "E",
    "UGU": "C", "UGC": "C", "UGA": "Z", "UGG": "W",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R",
    "AGU": "S", "AGC": "S", "AGA": "*", "AGG": "*",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G"}

cyt = {"UUU": "F", "UUC": "F", "UUA": "L", "UUG": "L",
    "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "AUU": "I", "AUC": "I", "AUA": "I", "AUG": "M",
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "UAU": "Y", "UAC": "Y", "UAA": "*", "UAG": "*",
    "CAU": "H", "CAC": "H", "CAA": "Q", "CAG": "Q",
    "AAU": "N", "AAC": "N", "AAA": "K", "AAG": "K",
    "GAU": "D", "GAC": "D", "GAA": "E", "GAG": "E",
    "UGU": "C", "UGC": "C", "UGA": "*", "UGG": "W",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R",
    "AGU": "S", "AGC": "S", "AGA": "R", "AGG": "R",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G"}
    

#defining functions

#returns the complementary strand in the correct order for translation (5' -> 3')
def reverse_complement(sequence):
    # Define a dictionary for complementing nucleotides
    complement = {'A': 'U', 'T': 'A', 'C': 'G', 'G': 'C','N':'N'}
    # Reverse the sequence
    reversed_sequence = sequence[::-1]
    # Complement each nucleotide and join them to form the reverse complement
    reverse_complement = ''.join([complement[nucleotide] for nucleotide in reversed_sequence])
    return reverse_complement

#DOUBLE TRIPLE CHECK MAKE SURE IT WORKS
def generate_ORFs(seq, start_codons, stop_codons, filter_cod):
    "Cette fonction fait chier."
    total_ORF = []
    for i in range(len(seq)-3):
        codon = seq[i:i+3]
        if codon in start_codons:
            #initiate ORF
            ORF = ''
            for nuc in range(i,len(seq),3):
                codon = seq[nuc:nuc+3]
                if codon not in stop_codons:
                    ORF += codon
                else:
                    ORF += codon
                    #minimal altORF length
                    if len(ORF)> filter_cod:
                        total_ORF.append(ORF)
                    break
    return total_ORF
              
def transcribe(dna_seq):
    
    rna_seq = dna_seq.replace("T", "U")
    return rna_seq
        

#need another translate function for mitochondrial (maybe have the codon table as input)
def translate(sequence, start_codons, stop_codons, codon_table):
   
    #make new code in which start and stop codons are derived from the codon table
    protein_sequence = ''
    i = 0
    while i < len(sequence):
        codon = sequence[i:i+3]
        #if codon_table[codon] == 'M'
        if codon in start_codons:
            protein_sequence += 'M'
        
        elif codon in stop_codons:
            protein_sequence += '*'
        elif codon in codon_table:
            protein_sequence += codon_table[codon]
        i += 3
    return protein_sequence

def translate_orfs(orfs,start,stop,codon_table):
    prot_orfs = []
    for i in range(len(orfs)):
        prot_orfs.append(translate(orfs[i],start,stop,codon_table))
    return prot_orfs

#turn a list of string into a list of seq objects to allow alignement
def seq_list(prots_list):
    newlist = []
    for seq in prots_list:
        newlist.append(Seq(seq))
        
    return newlist

# Define a list of protein sequences to compare against

def protein_aligner (query, list_of_prots):
    
    # Initialize a PairwiseAligner object
    aligner = PairwiseAligner()

    # Loop through the list of sequences to find the best match
    best_score = 0
    best_seq = None
    
    for seq in list_of_prots:
    # Use the aligner object to get the optimal global alignment between the two sequences
        alignments = aligner.align(query, seq)
    # Get the alignment score of the first alignment
        alignment_score = alignments[0].score
    # If the current alignment score is better than the previous best score, update the best score and best sequence
        if alignment_score > best_score:
            best_score = alignment_score
            best_seq = seq
    
    return best_score

#print('best match:   ',best_seq)
#print('query peptide:',query)
#print('match score:', best_score)
```
